# **About the Company**

In 2016, Cyclistic launched a successful bike-share offering. Since then, the program has grown to a fleet of 5,824 bicycles that
are geotracked and locked into a network of 692 stations across Chicago. The bikes can be unlocked from one station and
returned to any other station in the system anytime.
Until now, Cyclistic’s marketing strategy relied on building general awareness and appealing to broad consumer segments.
One approach that helped make these things possible was the flexibility of its pricing plans: single-ride passes, full-day passes,
and annual memberships. Customers who purchase single-ride or full-day passes are referred to as casual riders. Customers
who purchase annual memberships are Cyclistic members.
Cyclistic’s finance analysts have concluded that annual members are much more profitable than casual riders. Although the
pricing flexibility helps Cyclistic attract more customers, Moreno believes that maximizing the number of annual members will
be key to future growth. Rather than creating a marketing campaign that targets all-new customers, Moreno believes there is a
very good chance to convert casual riders into members. She notes that casual riders are already aware of the Cyclistic
program and have chosen Cyclistic for their mobility needs.
Moreno has set a clear goal: Design marketing strategies aimed at converting casual riders into annual members. In order to
do that, however, the marketing analyst team needs to better understand how annual members and casual riders differ, why
casual riders would buy a membership, and how digital media could affect their marketing tactics. Moreno and her team are
interested in analyzing the Cyclistic historical bike trip data to identify trends.

#### Data Manipulation



```r
# Change column names from "member_casual" to "usertype"

df <- df %>% 
rename(usertype = member_casual)
colnames(df)
table(df$usertype)
```

```r
head(df)
tail(df)
```

```r
dim(df)
```

```r
# aggregate ride data for each month, day, or year and create new column "day_of_week"
df$date <- as.Date(df$started_at)
df$month <- format(as.Date(df$date), "%m")
df$day <- format(as.Date(df$date), "%d")
df$year <- format(as.Date(df$date), "%Y")
df$day_of_week <- format(as.Date(df$date), "%A")
```

```r
head(df)
```

```r
tail(df)
```

```r
str(df)
```

```r
# Calculate trip duration in seconds and store as "ride_length"

df$ride_length <- difftime(df$ended_at,df$started_at)
head(df)
str(df)
```

```r
# Convert "ride_length" from difftime to numeric so we can run calculations on the data

is.difftime(df$ride_length)
df$ride_length <- as.numeric(as.difftime(df$ride_length))
is.numeric(df$ride_length)
str(df)
```

```r
# check for negative values in ride_length column 
filter(df, ride_length == "-7745")
```

```r
# Remove "bad" data
# The dataframe includes a few hundred negative entries when bikes were taken out of docks 
# We will create a new version of the dataframe (v2) since data is being removed

df_v2 <- df[!(df$start_station_name == "HQ QR" | df$ride_length<0),]
```

```r
#negative values in ride_length column have been eliminated which is 72 rows of bad data
filter(df_v2, ride_length == "-7745")
dim(df_v2)
```

### Analyze

```r
# desciptive analysis on ride length (in seconds)

summary(df_v2$ride_length)
```

```r
# compare casual vs member usertype using summary statistics 

aggregate(df_v2$ride_length ~ df_v2$usertype, FUN = mean)
aggregate(df_v2$ride_length ~ df_v2$usertype, FUN = median)
aggregate(df_v2$ride_length ~ df_v2$usertype, FUN = max)
aggregate(df_v2$ride_length ~ df_v2$usertype, FUN = min)
```

```r
aggregate(df_v2$ride_length ~ df_v2$usertype, FUN = median)
```

```r
aggregate(df_v2$ride_length ~ df_v2$usertype, FUN = max)
```

```r
aggregate(df_v2$ride_length ~ df_v2$usertype, FUN = min)
```

```r
# Average ride time by each day for members vs casual users
aggregate(df_v2$ride_length ~ df_v2$usertype + df_v2$day_of_week, FUN = mean)
```

```r
# Notice that the days of the week are out of order. Let's fix that.
df_v2$day_of_week <- ordered(df_v2$day_of_week, levels=c("Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday))


# Now, let's run the average ride time by each day for members vs casual users

aggregate(df_v2$ride_length ~ df_v2$usertype + df_v2$day_of_week, FUN = mean)

```

```r
# analyze ridership data by type and weekday
df_v2 %>% 
  mutate(weekday = wday(started_at)) %>%  #creates weekday field using wday()
  group_by(usertype, weekday) %>%  #groups by usertype and weekday
  summarise(number_of_rides = n()							#calculates the number of rides 
  ,average_duration = mean(ride_length)) %>% 		# calculates the average duration
  arrange(usertype, weekday)								# sorts
```

```r
# Analyzed Dataset
analyzed_df <- df_v2 %>% 
  group_by(usertype, day_of_week) %>% 
  summarise(number_of_rides = n(), average_duration = mean(ride_length)) %>%
arrange(usertype, day_of_week)


head(analyzed_df)
tail(analyzed_df)
```

```r
# Let's visualize the usertype proportion
ggplot (data = df_v2) +
geom_bar (mapping = aes(x = usertype, fill=usertype)) +
labs(title = "Usertype Proportion: Member vs Casual riders")
```

```r
# Let's visualize the usertype proportion by rideable bike type

ggplot (data = df_v2) +
geom_bar (mapping = aes(x = usertype, fill=usertype)) +
facet_wrap(~rideable_type)
labs(title = "Usertype Proportion: Member vs Casual riders")
```

```r
# Let's visualize the number of rides by usertype
df_v2 %>% 
  mutate(weekday = wday(started_at)) %>% 
  group_by(usertype, day_of_week) %>% 
  summarise(number_of_rides = n()
            ,average_duration = mean(ride_length)) %>% 
  arrange(usertype, day_of_week)  %>% 
  ggplot(aes(x = day_of_week, y = number_of_rides, fill = usertype)) +
  geom_col(position = "dodge") +
   labs(title = "number of rides by Usertype and Day of Week")
```

```r
# Let's create a visualization for average trip duration grouping by day of week and usertype
df_v2 %>% 
  mutate(weekday = wday(started_at)) %>% 
  group_by(usertype, day_of_week) %>% 
  summarise(number_of_rides = n()
            ,average_duration = mean(ride_length)) %>% 
  arrange(usertype, day_of_week)  %>% 
  ggplot(aes(x = day_of_week, y = average_duration, fill = usertype)) +
  geom_col(position = "dodge") +
  labs(title = "Average trip duration by Usertype and Day of Week")
```

#### Key Findings From visualization 

* Casual riders have about twice as much average trip duration as member riders, especially on Saturday and Sunday. The high average trip duration of casual riders on weekends might be caused by the use of rides for touristic or leisure activities.

* On weekdays, there are more number of rides by member riders whereas more number of rides by casual riders is recorded on weekends. I assume that more member riders use Cyclistic bikes to travel to work/school while casual riders use Cyclistic bikes for leisure purpose on weekends.

* More than half of the Cyclistic customer have signed up as annual member. This proved that the company is already building up a certain level of loyalty within its customer base.

* Majority of Member and Casual Cyclistic bike riders use classic bikes and few use docked bikes. 

* Of the 30% of Cyclistic bike users that employ her bike services to commute to work everyday, findings show that majority of those are members.

Based on the analysis I performed, I would love to suggest a few recommendations to convert more casual riders into annual memebers:
