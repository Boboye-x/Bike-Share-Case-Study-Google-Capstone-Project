One approach that helped make these things possible was the flexibility of its pricing plans: single-ride passes, full-day passes,

who purchase annual memberships are Cyclistic members.
Cyclistic’s finance analysts have concluded that annual members are much more profitable than casual riders. Although the

Moreno has set a clear goal: Design marketing strategies aimed at converting casual riders into annual members. In order to

#### Data Manipulation



```r
# create new column "ride_distance" which calculates the ride distances
df <- df %>% 
mutate(ride_distance = distHaversine(cbind(start_lng, start_lat), cbind(end_lng, end_lat)))
head(df)
```

```r
# Change column names from "member_casual" to "usertype"

df <- df %>% 
rename(usertype = member_casual)
colnames(df)
table(df$usertype)
```

```r
head(df)
tail(df)
```

```r
dim(df)
```

```r
# aggregate ride data for each month, day, or year and create new column "day_of_week"
df$date <- as.Date(df$started_at)
df$month <- format(as.Date(df$date), "%m")
df$day <- format(as.Date(df$date), "%d")
df$year <- format(as.Date(df$date), "%Y")
df$day_of_week <- format(as.Date(df$date), "%A")
head(df)
tail(df)
```

```r
str(df)
```

```r
# Calculate trip duration in seconds and store as "ride_length"

df$ride_length <- difftime(df$ended_at,df$started_at)
head(df)
str(df)
```

```r
# Convert "ride_length" from difftime to numeric so we can run calculations on the data

is.difftime(df$ride_length)
df$ride_length <- as.numeric(as.difftime(df$ride_length))
is.numeric(df$ride_length)
str(df)
```

```r
# removal of rides that are lesser than 0 seconds and 0 meters
df <- df %>% 
filter(ride_length > 0) %>%
filter(ride_distance > 0)
head(df)
dim(df)
```

### Analyze 

#### *Key Tasks*

* Perform calculations.
* Identify trends and relationships.
* Conduct descriptive analysis.

```r
# desciptive analysis on ride length (in seconds)

summary(df$ride_length)
```

```r
# compare ride_length of casual riders vs member usertype using summary statistics 

aggregate(df$ride_length ~ df$usertype, FUN = mean)
aggregate(df$ride_length ~ df$usertype, FUN = median)
aggregate(df$ride_length ~ df$usertype, FUN = max)
aggregate(df$ride_length ~ df$usertype, FUN = min)
```

```r
# compare ride_distance of casual riders vs member usertype using summary statistics 
aggregate(df$ride_distance ~ df$usertype, FUN = mean)
aggregate(df$ride_distance ~ df$usertype, FUN = median)
aggregate(df$ride_distance ~ df$usertype, FUN = max)
aggregate(df$ride_distance ~ df$usertype, FUN = min)
```

```r
# Average ride length by day_of_week for members vs casual users
aggregate(df$ride_length ~ df$usertype + df$day_of_week, FUN = mean)

# Average ride distance by day_of_week for members vs casual users
aggregate(df$ride_distance ~ df$usertype + df$day_of_week, FUN = mean)
```

```r
# Notice that the days of the week are out of order. Let's fix that.
df$day_of_week <- ordered(df$day_of_week, levels=c("Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"))


# Now, let's run the average ride length by day of week for members vs casual users

aggregate(df$ride_length ~ df$usertype + df$day_of_week, FUN = mean)

# Average ride distance by day_of_week for members vs casual users
aggregate(df$ride_distance ~ df$usertype + df$day_of_week, FUN = mean)
```

```r
head(df)
```

```r
# analyze ridership data by usertype and weekday
df %>% 
  mutate(weekday = wday(started_at)) %>%  #creates weekday field using wday()
  group_by(usertype, weekday) %>%  #groups by usertype and weekday
  summarise(number_of_rides = n()							#calculates the number of rides 
  ,average_duration = mean(ride_length)) %>% 		# calculates the average duration
  arrange(usertype, weekday)								# sorts
```

```r
# Analyzed Dataset
analyzed_df <- df %>% 
  group_by(usertype, day_of_week) %>% 
  summarise(number_of_rides = n(), average_duration = mean(ride_length),  average_distance = mean(ride_distance)) %>%
arrange(usertype, day_of_week)
as_tibble(analyzed_df)
```

From this Dataset we can infer that member riders have the highest number of rides during weekdays, peaking on tuesday while casual riders mostly ride during weekends peaking on saturday and  sunday.

```r
# Let's visualize the usertype proportion by rideable bike type

ggplot (data = df) +
geom_bar (mapping = aes(x = usertype, fill=usertype)) +
facet_wrap(~rideable_type)
labs(title = "Usertype Proportion: Member vs Casual riders")
```

```r
# Let's visualize the number of rides by usertype and day of week
df %>% 
  mutate(weekday = wday(started_at)) %>% 
  group_by(usertype, day_of_week) %>% 
  summarise(number_of_rides = n()) %>% 
  arrange(usertype, day_of_week)  %>% 
  ggplot(aes(x = day_of_week, y = number_of_rides, fill = usertype)) +
  geom_col(position = "dodge") +
   labs(title = "number of rides by Usertype and Day of Week")
```

```r
# Let's create a visualization for average trip duration grouping by day of week and usertype
df %>% 
  mutate(weekday = wday(started_at)) %>% 
  group_by(usertype, day_of_week) %>% 
  summarise(number_of_rides = n()
            ,average_duration = mean(ride_length)) %>% 
  arrange(usertype, day_of_week)  %>% 
  ggplot(aes(x = day_of_week, y = average_duration, fill = usertype)) +
  geom_col(position = "dodge") +
  labs(title = "Average trip duration by Usertype and Day of Week")
```

```r
df %>% 
  mutate(weekday = wday(started_at)) %>% 
  group_by(usertype, day_of_week) %>% 
  summarise(number_of_rides = n()
            ,average_duration = mean(ride_length), average_distance = mean(ride_distance)) %>% 
  arrange(usertype, day_of_week)  %>% 
  ggplot(aes(x = day_of_week, y = average_distance, fill = usertype)) +
  geom_col(position = "dodge") +
  labs(title = "Average trip distance by Usertype and Day of Week")
```

### Key findings from ggplot2 visualizations

* More than half of the Cyclistic customers have signed up as annual members. This proves that the company has already built up a certain level of loyalty among its customers. However, to increase revenue generation they hope to maximize the number of riders and target converting more casual riders to annual members.

* On weekdays, there are more rides by member riders peaking on Tuesday whereas more rides by casual riders are recorded on weekends. We can infer that more member riders use Cyclistic bikes to commute to work/school while casual riders use Cyclistic bikes for occasional or recreational purposes.

* Casual riders have about twice as much average trip duration as member riders, peaking on Saturday and Sunday. The high average trip duration of casual riders on weekends might be caused by the use of rides for touristic or leisure activities.

* The average trip distance for casual riders is also slightly longer than member riders, with both peaking on weekends.
